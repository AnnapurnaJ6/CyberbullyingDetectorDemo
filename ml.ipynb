```python
import pandas as pd #for processing rows and columns
import numpy as np#for processing numeric operations
import matplotlib.pyplot as plt #mathplotlib is library and pyplot is module inside the mathplotlib it used for the visualization of the data
from sklearn.linear_model import LinearRegression#sklearn is library linear_model is a module LinearRegression is class from which we are importing the models that best fits our logic
from sklearn.model_selection import train_test_split#sklearn is library model_selection is a module train_test_split is function
from sklearn.metrics import mean_squared_error,r2_score#sklearn is library metrics is a module mean_squared_error,r2_score is function to measure the accuracy
sheet_url="https://docs.google.com/spreadsheets/d/1FfnB6upRn_RhFdTGqYnKVtTzDud5Gv6W3yObodQwPIM/edit?usp=sharing"
csv_url=sheet_url.replace("/edit?usp=sharing","/export?format=csv")
df=pd.read_csv(csv_url)
print(df.head(60))

x=df[['Humidity','WindSpeed']]
y=df['Temperature']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

model=LinearRegression()
model.fit(x_train,y_train)

y_pred=model.predict(x_test)

print("Mean squared error:",mean_squared_error(y_test,y_pred))
print("R2 score:",r2_score(y_test,y_pred))

new_data=pd.DataFrame({'Humidity':[80],'WindSpeed':[10]})
pred_temp=model.predict(new_data)
print("predicted temperature:", pred_temp[0])
```
